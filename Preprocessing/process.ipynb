```python
import pandas as pd

data = pd.read_csv('data/orders.csv')
data.drop('order_id', axis = 1, inplace=True)
data.dropna(axis=0, inplace=True)
data.head()
```

```python
data.shape
```

```python
from datetime import datetime

estimated_delivery_date = data['order_estimated_delivery_date'].values
estimated_delivery_date = [datetime.strptime(i, '%Y-%m-%d %H:%M:%S') for i in estimated_delivery_date]

delivered_customer_date = data['order_delivered_customer_date'].values
delivered_customer_date = [datetime.strptime(i, '%Y-%m-%d %H:%M:%S') for i in delivered_customer_date]

data['order_estimated_delivery_date'] = estimated_delivery_date
data['order_delivered_customer_date'] = delivered_customer_date
```

```python
data['on_time'] = data['order_delivered_customer_date'] < data['order_estimated_delivery_date']
data['on_time'] = data['on_time'].replace({True:1, False:0})
```

# read customers.csv & sellers.csv

```python
customers = pd.read_csv('data/customers.csv')
customers.info()
customers.head()
```

```python
sellers = pd.read_csv('data/sellers.csv')
sellers.info()
sellers.head()
```

```python
data_customer = pd.merge(data, customers[['customer_id', 'customer_zip_code_prefix']], on = 'customer_id', how = 'left')
data_customer_seller = pd.merge(data_customer, sellers[['seller_id', 'seller_zip_code_prefix']], on = 'seller_id', how = 'left')
data_customer_seller.head(3)
```

```python
geolocation.rename(columns = {'geolocation_zip_code_prefix':'customer_zip_code_prefix'}, inplace=True)

data_customer_seller = pd.merge(data_customer_seller, geolocation[['customer_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']],
                                on='customer_zip_code_prefix', how='left')
data_customer_seller.head()
```

```python
data_customer_seller.rename(columns={'geolocation_lat': 'geolocation_lat_customer', 'geolocation_lng': 'geolocation_lng_customer'}, inplace=True)
data_customer_seller.head(3)
```

```python
geolocation.rename(columns = {'customer_zip_code_prefix':'seller_zip_code_prefix'}, inplace=True)

data_customer_seller = pd.merge(data_customer_seller, geolocation[['seller_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']],
                                on='seller_zip_code_prefix', how='left')
data_customer_seller.rename(columns={'geolocation_lat': 'geolocation_lat_seller', 'geolocation_lng': 'geolocation_lng_seller'}, inplace=True)
data_customer_seller.head()
```

# purchase, approved, carrier

```python
time_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

for column in time_columns:
    data_customer_seller.loc[:, column] = pd.to_datetime(data_customer_seller[column])
```

```python
data_customer_seller['wait_time'] = round((data_customer_seller['order_delivered_customer_date'] - data_customer_seller['order_purchase_timestamp']).dt.total_seconds()/3600, 2)
data_customer_seller['response_time'] = round((data_customer_seller['order_approved_at'] - data_customer_seller['order_purchase_timestamp']).dt.total_seconds()/60/60, 2)
data_customer_seller['prepare_time'] = round((data_customer_seller['order_delivered_carrier_date'] - data_customer_seller['order_approved_at']).dt.total_seconds()/3600, 2)
```

```python
data_customer_seller.head()
```

```python
data = data_customer_seller.drop(columns=['customer_id', 'order_status', 'order_purchase_timestamp', 'order_approved_at',
                                          'order_delivered_carrier_date','order_delivered_customer_date', 'order_estimated_delivery_date'
                                          ,'seller_id', 'customer_zip_code_prefix', 'seller_zip_code_prefix'])
data.head()
```

```python
import numpy as np

data['distance'] = np.sqrt((data['geolocation_lat_customer'] - data['geolocation_lat_seller'])**2 + (data['geolocation_lng_customer'] - data['geolocation_lng_seller'])**2)
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.shape
```

```python
X = data.loc[:, data.columns != 'on_time'].values
y = data['on_time'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=36)
```

```python
import pickle

with open('X_train.pickle', 'wb') as fp:
    pickle.dump(X_train, fp)
with open('X_test.pickle', 'wb') as fp:
    pickle.dump(X_test, fp)
with open('y_train.pickle', 'wb') as fp:
    pickle.dump(y_train, fp)
with open('y_test.pickle', 'wb') as fp:
    pickle.dump(y_test, fp)
```
