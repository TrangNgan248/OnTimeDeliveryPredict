```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
test_size = 0.2
random_state = 42
```

## **Dataset 1**

Nguồn: https://www.kaggle.com/datasets/prachi13/customer-analytics?fbclid=IwAR2yPwk6GvEmjkcmgul9g0fVUvgVOZsuH6JA2mc9oyrdcinUbvRmwduUr0c

```python
#Đọc dữ liệu từ file Data1.csv

df_1 = pd.read_csv("Data1.csv")
df_1.head()
```

Giải thích ý nghĩa các trường trong tập dữ liệu
*   ***ID***:
*   ***Warehouse_block***:
*   Mục danh sách
*   Mục danh sách
*   Mục danh sách
*   Mục danh sách







Kiểm tra định dạng, dữ liệu null của các trường dữ liệu

```python
df_1.info()
```

Chuyển các dữ liệu định dạng object về dạng số

Sử dụng phương thức LabelEncoder có sẵn trong thư viện scikit-learn để gán một số nguyên duy nhất cho mỗi giá trị có định dạng object

```python
le = LabelEncoder()
is_Category = df_1.dtypes == object
category_column_list = df_1.columns[is_Category].tolist()
df_1[category_column_list] = df_1[category_column_list].apply(lambda col: le.fit_transform(col))
```

```python
df_1.head()
```

```python
df_1.info()
```

Sau khi xử lý dữ liệu, sử dụng phương thức train_test_split có sẵn trong thư viện scikit-learn để chia tập dữ liệu thành tập huấn luyện và tập kiểm thử (với tỉ lệ 8:2)

```python
train, test = train_test_split(df_1, test_size=test_size, random_state=random_state)
```

Lưu thành 2 file Data1_train.csv và Data1_test.csv

```python
train.to_csv('Data1_train.csv')
test.to_csv('Data1_test.csv')
```

# **Dataset 2**

Nguồn: https://www.kaggle.com/datasets/ramakrishnanthiyagu/delivery-truck-trips-data

```python
df_2 = pd.read_excel('Delivery truck trip data.xlsx')
df_2.head(3)
```

Giải thích ý nghĩa các trường trong bảng dữ liệu
*   Mục danh sách
*   Mục danh sách
*   Mục danh sách
*   Mục danh sách





```python
df_2.info()
```

Thực hiện tiền xử lý như sau:

*   Các trường Minimum_kms_to_be_covered_in_a_day, Driver Mobil No, Drive name đều có quá nửa bản ghi là NULL, vì vậy ta sẽ bỏ đi 3 trường này.
*   Trường ontime, delay để dự đoán thực chất là một. Ta chuyển giá trị "G" --> 1 và NULL --> 0 ở trường ontime, đồng thời bỏ trường delay.

*  4 trường Current_Location, Curr_lat, Curr_lon, Data_Ping_time dùng để chỉ thông tin vị trí hiện tại của người vận chuyển khi mà nhận được tín hiệu từ GPS.
Do ta thực hiện việc dự đoán đúng giờ hay không trước khi đơn đặt hàng được vận chuyển nên ta sẽ bỏ 4 trường này.
*   Trường BookingID và BookingID_date, trip_end_date, actual_eta không có ý nghĩa trong việc dự đoán. Ta sẽ bỏ đi trường này.

*   Trường DestinationLocation bị trùng với trường Destination_location vậy ta sẽ bỏ đi một trong hai trường này.

*   2 trường customerNameCode, supplierNameCode chỉ là tên khách hàng và nhà cung cấp có ID ở các trường customerID và supplierID nên ta sẽ bỏ đi hai trường này.


*   Tách 2 trường Org_lat_lon và Des_lat_lon thành các trường kinh độ, vĩ độ riêng.

*   Ta sẽ bỏ đi dữ liệu thời gian mà chỉ giữ lại ngày trong trường trip_start_date và Planned_ETA









```python
#Bỏ các trường không cần thiết
df_2.drop(axis = 1, columns = ["Minimum_kms_to_be_covered_in_a_day", "Driver_MobileNo", "Driver_Name", "delay", "Current_Location", "Curr_lat", "Curr_lon", "Data_Ping_time",
                              "BookingID", "BookingID_Date", "DestinationLocation", 'trip_end_date', 'actual_eta', "customerNameCode", "supplierNameCode"], inplace = True)

#Thay G = 1, Null = 0 trong trường dự đoán ontime
df_2.ontime.replace({'G':1, np.NaN:0}, inplace=True)

#Tách kinh độ, vĩ độ thành 2 trường khác nhau
df_2[['Org_lat', 'Org_lon']] = df_2['Org_lat_lon'].str.split(',', expand=True).astype(float)
df_2[['Des_lat', 'Des_lon']] = df_2['Des_lat_lon'].str.split(',', expand=True).astype(float)
df_2.drop(axis = 1, columns = ["Org_lat_lon", "Des_lat_lon"], inplace = True)

#Bỏ đi dữ liệu thời gian, giữ lại dữ liệu ngày
df_2['trip_start_date'] = df_2['trip_start_date'].astype(str).str.split(' ', expand=True)[0]
df_2['Planned_ETA'] = df_2['Planned_ETA'].astype(str).str.split(' ', expand=True)[0]

#Bỏ đi bản ghi có giá trị NULL
df_2.dropna(inplace = True)
```

```python
df_2.info()
```

```python
df_2.isnull().sum()
```

Chuyển các dữ liệu định dạng object về dạng số

Sử dụng phương thức LabelEncoder có sẵn trong thư viện scikit-learn để gán một số nguyên duy nhất cho mỗi giá trị có định dạng object

```python
le = LabelEncoder()
is_Category = df_2.dtypes == object
category_column_list = df_2.columns[is_Category].tolist()
df_2[category_column_list] = df_2[category_column_list].astype(str).apply(lambda col: le.fit_transform(col))
```

Sau khi xử lý dữ liệu, sử dụng phương thức train_test_split có sẵn trong thư viện scikit-learn để chia tập dữ liệu thành tập huấn luyện và tập kiểm thử (với tỉ lệ 8:2)

```python
train, test = train_test_split(df_2, test_size=test_size, random_state=random_state)
```

Lưu lại thành hai file Data2_train và Data2_test

```python
train.to_csv("Data2_train.csv")
test.to_csv("Data2_test.csv")
```

# **Dataset 3**

Nguồn: https://www.kaggle.com/datasets/devdataanalyst/food-delivery-performance

```python
df_3 = pd.read_excel("Food delivery Performance.xlsx")
```

```python
df_3.head()
```

```python
df_3.info()
```

Thực hiện tiền xử lí như sau:
*   Bỏ các trường reason, sub_reason, owner do không liên quan và thiếu quá nhiều dữ liệu.
*   Bỏ các trường order_id, analytical_customer_id do không liên quan.
*   Bỏ trường gmv_amount_lc, do bị thừa vì gmv_amount_lc = deliver_fee + basket_amount
*   Tạo trường mới là ontime nhận giá trị 1 nếu order_delay < 0 và nhận giá trị 0 nếu ngược lại. Đồng thời xóa các trường actual_delivery_time, promised_delivery_time, order_delay.

*   Bỏ dữ liệu thời gian, giữ lại dữ liệu ngày trong trường order_time







```python
df_3.drop(axis = 1, columns = ["reason", "sub_reason", "owner", "order_id","analytical_customer_id", "gmv_amount_lc"], inplace=True)

df_3['ontime'] = df_3['order_delay'].apply(lambda x: 1 if x < 0 else 0)
df_3.drop(axis = 1, columns = ["actual_delivery_time", "promised_delivery_time", "order_delay"], inplace=True)

#Bỏ đi dữ liệu thời gian, giữ lại dữ liệu ngày
df_3['order_time'] = df_3['order_time'].astype(str).str.split(' ', expand=True)[0]

df_3.dropna(axis = 1, inplace = True)
```

```python
df_3.info()
```

Chuyển các dữ liệu định dạng object về dạng số

Sử dụng phương thức LabelEncoder có sẵn trong thư viện scikit-learn để gán một số nguyên duy nhất cho mỗi giá trị có định dạng object


```python
le = LabelEncoder()

is_bool = df_3.dtypes == bool
df_3[df_3.columns[is_bool].tolist()] = df_3[df_3.columns[is_bool].tolist()].astype(str)

is_Category = df_3.dtypes == object
category_column_list = df_3.columns[is_Category].tolist()
df_3[category_column_list] = df_3[category_column_list].astype(str).apply(lambda col: le.fit_transform(col))
```

```python
df_3.head()
```

Sau khi xử lý dữ liệu, sử dụng phương thức train_test_split có sẵn trong thư viện scikit-learn để chia tập dữ liệu thành tập huấn luyện và tập kiểm thử (với tỉ lệ 8:2)

```python
train, test = train_test_split(df_3, test_size=test_size, random_state=random_state)
```

Lưu tập train, test tên là Data3_train.csv và Data3_test.csv

```python
train.to_csv("Data3_train.csv")
test.to_csv("Data3_test.csv")
```
